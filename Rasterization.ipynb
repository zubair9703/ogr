```python
from osgeo import gdal,ogr,gdalconst
import os
import numpy as np
import ospybook as pb
```

```python
os.chdir('J:/New folder/')
```

```python
in_ds=ogr.Open('Bathymetry_IOR.geojson')
```

```python
in_ly=in_ds.GetLayer()
```

```python
# in_ly.SetAttributeFilter('DEPTH>= -1000  ')
```

```python
pb.print_attributes(in_ly)
```

```python
envelopes =[row.geometry().GetEnvelope() for row in in_ly]
```

```python
coords = list(zip(*envelopes)) 
print(coords)
```

```python
minx, maxx = min(coords[0]), max(coords[1]) 
miny, maxy = min(coords[2]), max(coords[3]) 
```

```python
cellsize=0.01
```

```python
tif_driver = gdal.GetDriverByName('GTiff')
cols = int((maxx - minx) / cellsize)
rows = int((maxy - miny) / cellsize)
```

```python
out_ds = tif_driver.Create('Bathymetry_rasterizedfull.tif', cols, rows,1,gdal.GDT_Int16, ['COMPRESS=LZW']) 
out_ds.SetProjection(in_ly.GetSpatialRef().ExportToWkt())
```

```python
out_ds.SetGeoTransform((minx, cellsize, 0, maxy, 0, -cellsize))
```

```python
gdal.RasterizeLayer(out_ds, [1], in_ly,  options=["ATTRIBUTE=DEPTH"])
```

```python
del out_ds
```
